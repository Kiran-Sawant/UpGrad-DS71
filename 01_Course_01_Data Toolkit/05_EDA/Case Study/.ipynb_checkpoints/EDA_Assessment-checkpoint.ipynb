## 4.3 Handling outliers and Null values

### 4.3.1 AMT_ANNUITY Imputation

```python
# All rows in applications where 'AMT_ANNUITY' is Null.
df_application[df_application['AMT_ANNUITY'].isnull()]
```

```python
df_application['AMT_ANNUITY'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.99])
```

```python
sns.boxplot(df_application["AMT_ANNUITY"], orient='h')
plt.style.use("ggplot")
plt.show()
```

By observing the descriptive statistics and the box plot we ccan clearly see that the 99th percentile is far off from the max value of 258025.5. Therefore, we may impute the Null values in this column with the median of 24903.

```python
# Storing the median in a variable.

annuity_median = df_application['AMT_ANNUITY'].median()
```

```python
# Replacing Null values with median

df_application['AMT_ANNUITY'].fillna(annuity_median, inplace=True)
```

```python
# Chacking for Null values after imputation

df_application['AMT_ANNUITY'].isnull().sum()
```

### 4.3.2 AMT_GOODS_PRICE Imputation

```python
# Observing rows where AMT_GOODS_PRICE is Null

df_application[df_application['AMT_GOODS_PRICE'].isnull()]
```

```python
# Descriptive Stats of AMT_GOODS_PRICE

df_application['AMT_GOODS_PRICE'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.99])
```

```python
sns.boxplot(df_application['AMT_GOODS_PRICE'], orient='h')
plt.show()
```

As we can see, There are many outliers in the *Goods Price* column. The max value is 4x the value at 99th percentile. We can impute median value in ull values. However, Prise of the good for which the loan is taken is a very important metric for consideration. As there are only 220 Null values missing values may not affect our final outcom. we may leave Null values as is.

### 4.3.3 NAME_TYPE_SUIT Imputation

- NAME_TYPE_SUIT: Person who was accompanying client when he was applying for the loan
- Missing values 1292 (0.42%)
- Variable type: Catagorical

```python
# Finding what values NAME_TYPE_SUITE consists of.

df_application['NAME_TYPE_SUITE'].unique()
```

```python
# Finding the counts of each unique value.

df_application['NAME_TYPE_SUITE'].value_counts()
```

```python
# Finding the mode and storing it in a variable

NAME_TYPE_SUITE_mode = df_application['NAME_TYPE_SUITE'].mode()

NAME_TYPE_SUITE_mode.iloc[0]
```

As this is a catagorical variable, we do not need to use mean and median. Instead we shall impute Null values with the mode that is "Unaccompained". Unaccompained consts of a majority of values i.e. it is 6x more thann the second most common value i.e. 'Family'

```python
# Replacing Null with the mode

df_application['NAME_TYPE_SUITE'].fillna("Unaccompanied", inplace=True)
```

```python
# Calculating Null count after imputation.

df_application['NAME_TYPE_SUITE'].isnull().sum()
```

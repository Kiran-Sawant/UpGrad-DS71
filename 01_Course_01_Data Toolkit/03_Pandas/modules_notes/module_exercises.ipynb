## Employee Training

You are provided with the dataset of a company which has offices across three cities - Mumbai, Bangalore and New Delhi. The dataset contains the rating (out of 5) of all the employees from different departments (Finance, HR, Marketing and Sales).<br><br>

The company has come up with a new policy that any individual with a rating equal to or below 3.5 needs to attend a training. Using dataframes, load the dataset and then derive the column ‘Training’ which shows ‘Yes’ for people who require training and ‘No’ for those who do not.

```python
df_emp = pd.read_csv('https://media-doselect.s3.amazonaws.com/generic/NMgEjwkAEGGQZBoNYGr9Ld7w0/rating.csv')

df_emp
```

```python
df_emp['Training'] = df_emp['Rating'].apply(lambda x: 'Yes' if x <= 3.5 else 'No')
```

```python
df_emp
```

```python
df_emp[(df_emp['Department'] == 'Finance') & (df_emp['Training'] == 'Yes')].info()
```

```python
df_emp[(df_emp['Department'] == 'Marketing') & (df_emp['Training'] == 'Yes')].info()
```

```python
df_emp[(df_emp['Department'] == 'Sales') & (df_emp['Training'] == 'Yes')].info()
```

```python
df_emp[(df_emp['Department'] == 'HR') & (df_emp['Training'] == 'Yes')].info()
```

## Dataframe grouping

Group the dataframe 'df' by 'month' and 'day' and find the mean value for column 'rain' and 'wind'.

```python
grp = df_fire.groupby(['month', 'day'])

grp[['rain', 'wind']].mean(numeric_only=True)
```

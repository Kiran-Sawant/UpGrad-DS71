## **Bar Charts**:
- **Purpose:** Bar charts are used to display and compare the frequency, count, or other measures (like average) for different categories.
- **Structure:**
  - **Bars:** Each category is represented by a bar. The height (or length, in the case of horizontal bars) of the bar corresponds to the value.
  - **Axes:**
    - **X-axis:** Usually represents the categories.
    - **Y-axis:** Usually represents the values.
  
### Basic Example:
```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Creating the bar chart
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')

# Displaying the chart
plt.show()
```

### Key Points:
- **Categories:** Represented along one axis (X-axis in the example).
- **Values:** Heights of bars represent the values (Y-axis in the example).
- **Customization:** You can customize colors, add labels, titles, grids, and more.

Bar charts are great for visualizing differences among categories and spotting trends in your data at a glance. They're simple yet powerful! ðŸ˜Š

#### Plot sales across each product category

1. Adding labels to Axes
2. Reducing the bar width
3. Giving Title to the chart
4. Modifying the ticks to show information in (million dollars)

````python
import numpy as np
import matplotlib.pyplot as plt

product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array([4110451.90, 4744557.50, 3787492.52])

# plotting bar chart and setting bar width to 0.5 and aligning it to center
# edgecolor will give a fine boarder with that color.
# color parameter can take a list of strings, one color for each bar. for additional bars, color repeats.
plt.bar(product_category, sales, width= 0.5, align='center', edgecolor='Orange', color='cyan')

# Adding and formatting title
plt.title("Sales Across Product Categories\n", fontdict={'fontsize': 20, 'fontweight': 5, 'color': 'Green'})

# Labeling Axes
plt.xlabel("Product Category", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})


# Modifying the ticks to show information in (million dollars)
"""ticks are numerical values used to draw the plot.
While labels are values displayed"""

ticks = np.arange(0, 6000000, 1000000)                      # a range of values for y-axis ticks
labels = list(f"{i//1000000}M" for i in ticks)              # a list of ticks to feed the .yticks() method
plt.yticks(ticks, labels)                                   # setting ticks and labels on the y-axis.

plt.show()
````


## **Scatter Plots**:
Scatter plots are fantastic for visualizing the relationship between two variables:<br>
- **Purpose:** Scatter plots are used to observe and show relationships or correlations between two numerical variables. Each point on the plot represents an observation. The keyword is versus, if you want to observe, one versus the other over a pair of parameters, use scatter plot.

### Example:
Consider a real-life use case where we want to analyze the relationship between the amount of advertising budget spent and sales revenue generated.

### Code Example:
```python
import matplotlib.pyplot as plt

# Sample data: Advertising budget (in thousands) and corresponding sales (in thousands)
advertising_budget = [10, 15, 20, 25, 30, 35, 40]
sales_revenue = [40, 45, 50, 55, 60, 65, 70]

# Creating a scatter plot
plt.scatter(advertising_budget, sales_revenue)

# Adding labels and title
plt.xlabel('Advertising Budget (in $1000)')
plt.ylabel('Sales Revenue (in $1000)')
plt.title('Advertising Budget vs. Sales Revenue')

# Displaying the plot
plt.show()
```

### Key Points:
- **X-axis:** Represents the advertising budget.
- **Y-axis:** Represents the sales revenue.
- **Points:** Each point represents an observation (a pair of advertising budget and corresponding sales revenue).

### Real-Life Use Case:
In this example, a company might want to understand how different levels of investment in advertising impact their sales. By plotting the advertising budget against sales revenue, they can visually inspect if there is a positive correlation (which there appears to be in the sample data) and use this information to make informed decisions on future advertising spend.

Scatter plots are powerful for identifying trends, clusters, and potential outliers in data. They are widely used in fields like finance, marketing, research, and many more.

## **Line Chart**

- **Structure:**
  - **X-axis:** Represents the independent variable (e.g., time).
  - **Y-axis:** Represents the dependent variable (e.g., sales, temperature).

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]

plt.plot(x, y)

# Adding labels and title

plt.ylabel('Y-axis')
plt.title('Simple Line Chart')

# Displaying the chart
plt.show()
```

### Key Points:
- **Continuous Data:** Best for data that changes continuously over a range.
- **Trends:** Excellent for visualizing trends, patterns, and changes over time.

The annotate() function in the pyplot module of Matplotlib is used to: 
- Draw an arrow that connects two points on a plot 
- Add a label text and arrow component to the appropriate axes 
- To use the annotate() function, you can:
- Call the annotate() function before the show() function<br><br>

- **label_text**: The text to add to the plot <br>
- **xy**: The tuple representing the x and y coordinates of the point where the arrow will start <br>



## **Box and Whisker Chart**

Box plots, also known as box-and-whisker plots, are a great way to summarize and visualize the distribution of a dataset.<br>They highlight the central tendency, variability, and presence of any outliers.<br>Hereâ€™s a brief overview:

### Key Components of a Box Plot:
1. **Median (Q2):** The middle value of the dataset, represented by a line inside the box.
2. **Quartiles (Q1 and Q3):** 
   - **Q1 (Lower Quartile):** The median of the lower half of the dataset (25th percentile).
   - **Q3 (Upper Quartile):** The median of the upper half of the dataset (75th percentile).
3. **Interquartile Range (IQR):** The range between Q1 and Q3, showing the middle 50% of the data.
4. **Whiskers:** Lines extending from the box to the minimum and maximum values within 1.5 * IQR from the quartiles.
5. **Outliers:** Data points outside the whiskers are considered outliers and are often marked as individual points.

![Box](\images\Boxplot.png)

### Example in Matplotlib:
```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(10)
data = np.random.normal(0, 1, 100)

# Creating the box plot
plt.boxplot(data)

# Adding labels and title
plt.xlabel('Sample Data')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Displaying the plot
plt.show()
```

### Key Points:
- **Box:** Represents the IQR (middle 50% of the data). The edges of the box are Q1 and Q3.
- **Line in the Box:** Represents the median (Q2).
- **Whiskers:** Extend to the smallest and largest values within 1.5 * IQR from Q1 and Q3, respectively.
- **Outliers:** Values beyond the whiskers, shown as individual points.

### Use Cases:
- **Comparing Distributions:** Box plots are useful for comparing distributions across different groups or categories.
- **Identifying Outliers:** They highlight outliers and variability in the dataset.
- **Summarizing Data:** Provide a quick visual summary of the data's central tendency and spread.

Box plots are widely used in exploratory data analysis (EDA) to understand data distributions and identify any anomalies.

- **Histogram**: Used to represent the distribution of continuous data. The data is divided into bins, and the bars show the frequency of data points within each bin.<br>
- **Bar Chart**: Used to compare categorical data. Each bar represents a category, and the height of the bar shows the value or frequency of that category.

## **SubPlots**

````python
import numpy as np
import matplotlib.pyplot as plt

# Year array
years = np.array(['2012', '2013', '2014', '2015'])

# Sales Arrays
sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
````

### Subplots are shown in the same graph as line charts, identified by different colours

````python
# Figure is the entire chart, axis are individual plots on it.
"""A thing to note is that each instance of ax must be declared followed by a comma"""
fig, ax = plt.subplots()

europe, = ax.plot(years, sales_Europe)
europe.set_label("Europe")

usca, = ax.plot(years, sales_USCA)
usca.set_label("USCA")
usca.set_dashes([2, 2, 2, 2])                   # 2-point line, 2-point break

africa, = ax.plot(years, sales_africa)
africa.set_label("Africa")
africa.set_dashes([2, 2, 5, 2])                 # 2-point line, 2-point break, 5-point line, 2-point break

asia, = ax.plot(years, sales_Asia_Pacific)
asia.set_label("Asia Pacific")
asia.set_dashes([2, 2, 10, 2])                  # 2-point line, 2-point break, 10-point line, 2-point break

latam, = ax.plot(years, sales_LATAM)
latam.set_label("LATAM")
latam.set_dashes([2, 5, 5, 2])                  # 2-point line, 5-point break, 5-point line, 2-point break

# legend to differentiate each line
plt.legend()
plt.legend(bbox_to_anchor=(1.31,0.4))

plt.show()
````

### Subplots shown in seperate graphs with each chart of a different types.

- `plt.subplot(nrow, ncol, x)`: It creates a subplot. 'nrow' and 'ncol' are the dimensions of the array inside the figure, and 'x' is the position in the array.
(You can visit this web page to understand the attributes associated with the subplot method in detail.)

- In `plt.subplot()` method, the numbering of subplots starts from the top-left element of the grid and moves rightward along each row. The numbering then continues to the next row from left to right. For example, suppose you have created a grid of nine subplots. The numbering would look like this:

|Subplot 1|	Subplot 2|	Subplot 3|
|---------|----------|-----------|
|**Subplot 4**|**Subplot 5**|**Subplot 6**|
|**Subplot 7**|**Subplot 8**|**Subplot 9**|

Suppose you need to access the subplot present in the second row and the second column (i.e., 'Subplot 5' in the table above). You can do so by using the following command:<br>
`plt.subplot(3, 3, 5)`

````python
# initiating a plot with 3 columns and 2 rows sharing x and y axis
fig, ax = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True)

europe, = ax[0][0].plot(years, sales_Europe)            # line plot at 1st row, 1 column
europe.set_label("Europe")

usca = ax[0][1].bar(years, sales_USCA)                  # Bar chart at 1st row, 2nd column
usca.set_label("USCA")

africa = ax[0][2].scatter(years, sales_africa)          # Scatter chart at 1st row, 3rd column
africa.set_label("Africa")

asia, = ax[1][0].plot(years, sales_Asia_Pacific)
asia.set_label("Asia Pacific")
asia.set_dashes([2, 2, 10, 2])

latam, = ax[1][1].plot(years, sales_LATAM)
latam.set_label("LATAM")
latam.set_dashes([2, 5, 5, 2])


plt.show()
````

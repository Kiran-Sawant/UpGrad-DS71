## Histogram

The difference between a bar chart and a histogram is that the values are in the x axis adn they-axis shows the number of occurances of the value within a certain range i.e. bin<br>
A bin is the short range on the x axis of the histogram. The bar right above a bin represents the number of values in the data within that bins range.

```python
profit = np.array(list(ran.choice(range(-20000, 60000)) for i in range(10000)))

plt.hist(profit, bins=100, edgecolor='green', color='yellow')
plt.show()
```

## Box plot

Box plots are quite effective in summarising the spread of a large data set into a visual representation. They use percentiles to divide the data range.<br><br>

The percentile value gives the proportion of the data range that falls below a chosen data point when all the data points are arranged in the descending order. For example, if a data point with a value of 700 has a percentile value of 99% in a data set, then it means that 99% of the values in the data set are less than 700.<br><br>

Box plots divide the data range into three important categories, which are as follows:
-  Median value: This is the value that divides the data range into two equal halves, i.e., the 50th percentile.
-  Interquartile range (IQR): These data points range between the 25th and 75th percentile values.
-  Outliers: These are data points that differ significantly from other observations and lie beyond the whiskers.
